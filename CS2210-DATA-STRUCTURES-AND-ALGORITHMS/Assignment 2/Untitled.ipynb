```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
hashtable = ((1000, 57),(2000,70),(4000,88),(6000,103),(8000,135),(10000,115)) #x is number of words, y is runtime
hx = [1000,2000,4000,6000,8000,10000]
hy = [57,70,88,103,135,115]

linkedlist = ((1000, 75),(2000,105),(4000,168),(6000,230),(8000,340),(10000,475))
lx = [1000,2000,4000,6000,8000,10000]
ly = [75,105,168,230,340,475]
```

```python
plt.figure(figsize = (15,10))
plt.plot(hx,hy, label='hashtable')
plt.plot(lx,ly, label='linkedlist')
plt.xlabel("Number of words")
plt.ylabel("Total Runtime (ms)")
plt.legend()
plt.savefig("comparison_chart.pdf")
```

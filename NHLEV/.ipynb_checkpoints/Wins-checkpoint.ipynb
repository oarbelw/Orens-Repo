```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn as sk
import sklearn.linear_model as skl
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, LabelEncoder, OneHotEncoder
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_selection import SelectFromModel
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LassoCV, ElasticNetCV
import matplotlib.pyplot as plt
from IPython.display import display
from itertools import groupby
```

```python
df = pd.read_csv("teams.csv")
df = df.drop("name", axis='columns')
df = df.drop("team.1", axis='columns')
df = df.drop("position", axis='columns')

df.head()
```

```python
situations = pd.unique(df["situation"]).tolist()
teams = pd.unique(df["team"]).tolist()

cats = df.columns.values[4:]
newdf = pd.DataFrame()
newdf["Team"] = teams

i = 0
for sit in situations:
    tempdf = pd.DataFrame(columns=[str(sit)+"_"+str(cat) for cat in cats])
    j=i
    k=0
    while (j < df.shape[0]):
        tempdf.loc[k] = df.iloc[j].values[4:]
        k+=1
        j+=5
    i+=1
    newdf = pd.concat([newdf, tempdf], axis=1)

newdf["Team"] = teams
print(newdf.shape)
newdf = newdf.sort_values(by = ['Team'])
newdf = newdf.reset_index(drop=True)
# newdf
```

```python
# Library for opening url and creating
# requests
import urllib.request
 
# pretty-print python data structures
from pprint import pprint
 
# for parsing all the tables present
# on the website
from html_table_parser.parser import HTMLTableParser
 
# for converting the parsed data in a
# pandas dataframe
import pandas as pd
 
 
# Opens a website and read its
# binary contents (HTTP Response Body)
def url_get_contents(url):
 
    # Opens a website and read its
    # binary contents (HTTP Response Body)
 
    #making request to the website
    req = urllib.request.Request(url=url)
    f = urllib.request.urlopen(req)
 
    #reading contents of the website
    return f.read()
 
url = "https://www.covers.com/sport/hockey/nhl/standings"
# defining the html contents of a URL.
xhtml = url_get_contents(url).decode('utf-8')
 
# Defining the HTMLTableParser object
p = HTMLTableParser()
 
# feeding the html contents in the
# HTMLTableParser object
p.feed(xhtml)
 
# Now finally obtaining the data of
# the table required
# pprint(p.tables[0])
 
# converting the parsed data to
# dataframe
# print("\n\nPANDAS DATAFRAME\n")
colnames = p.tables[0][1][1:]
colnames.insert(0,"Team")
standingsdf = pd.DataFrame(p.tables[0], columns = [str(col) for col in colnames])
# dataOHE = dataOHE.drop([col], axis=1)

# standingsdf
```

```python
NHLabvrs = {
"Anaheim" : "ANA",
"Atlanta" : "ATL",
"Boston" : "BOS",
"Buffalo" : "BUF",
"Carolina" : "CAR",
"Columbus" : "CBJ",
"Calgary" : "CGY",
"Chicago" : "CHI",
"Colorado" : "COL",
"Dallas" : "DAL",
"Detroit" : "DET",
"Edmonton" : "EDM",
"Florida" : "FLA",
"Los Angeles" : "LA",
"Minnesota" : "MIN",
"Montreal" :"MTL",
"Nashville" : "NSH",
"New Jersey" : "NJ",
"NY Islanders" : "NYI",
"NY Rangers" : "NYR",
"Ottawa" : "OTT",
"Philadelphia" : "PHI",
"Arizona" : "ARI",
"Pittsburgh" : "PIT",
"San Jose" : "SJ",
"St. Louis" : "STL",
"Tampa Bay" : "TB",
"Toronto" : "TOR",
"Vancouver" : "VAN",
"Winnipeg" : "WPG",
"Washington" : "WSH",
"Seattle" : "SEA",
"Vegas" : "VGK",
}
```

```python
for teamname in standingsdf["Team"]:
#     print(colname)
    changed = False
    for el in NHLabvrs:
        if el == teamname:
            standingsdf['Team'] = standingsdf['Team'].replace([teamname], NHLabvrs[el])
            changed = True
    if changed == False:
        ind = standingsdf.index[standingsdf["Team"] == teamname].tolist()
        standingsdf = standingsdf.drop(index=ind)
standingsdf = standingsdf.sort_values(by = ['Team'])
standingsdf = standingsdf.reset_index(drop=True)
# standingsdf
```

```python
df = pd.concat([newdf["Team"], standingsdf["W"], newdf.iloc[:,1:]], axis=1)
df
```

```python
X = df.drop(["W", "Team"], axis='columns')
y = df["W"]
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.30)
print("number of observations in the training data set: ", Xtrain.shape[0])
print("average value of the target variable in the training data set: ", np.mean(ytrain))
```

```python
linreg = skl.LinearRegression(fit_intercept=False)
linreg.fit(Xtrain,ytrain)
```

```python
selector = SequentialFeatureSelector(estimator=linreg,
                                     direction='forward')
selector.fit(Xtrain,ytrain)
selected_features = selector.get_support(indices=True)
```

```python
Xtrain_sel = Xtrain.iloc[:,selected_features]
# print(len(Xtrain_sel))
Xtest_sel = Xtest.iloc[:,selected_features]
linreg_sel = linreg.fit(Xtrain_sel, ytrain)
ypred = linreg_sel.predict(Xtest_sel)
# print(ypred)
inds = Xtest_sel.index.values
# resdf = pd.DataFrame()
for i in range(len(inds)):
    tm = df.iloc[inds[i]].values[0]
    wins = df.iloc[inds[i]].values[1]
    win_pred = ypred[i]
    print("{} has {} wins, predicted to have {:.3f}".format(tm,wins,win_pred))

# for i in range(len(df)):
#     dftmp = df.loc[i:i]
#     tX = dftmp.drop(["W", "Team"], axis='columns')
#     ty = dftmp["W"]
#     tXtrain_sel = tX.iloc[:,selected_features]
#     yt = linreg_sel.predict(tXtrain_sel)
#     tm = dftmp.iloc[0].values[0]
#     wins = dftmp.iloc[0].values[1]
#     win_pred = yt
#     print(win_pred)
#     print("{} has {} wins, predicted to have {}".format(tm,wins,win_pred))
# #     print(yt)

# ypredtrain = linreg_sel.predict(Xtrain_sel)
# indstrain = Xtrain_sel.index.values
# for i in range(len(indstrain)):
#     tm = df.iloc[indstrain[i]].values[0]
#     wins = df.iloc[indstrain[i]].values[1]
#     ypredtrain = linreg_sel.predict(wins)
#     win_pred = ypredtrain[i]
#     print("{} has {} wins, predicted to have {:.3f}".format(tm,wins,win_pred))
```
